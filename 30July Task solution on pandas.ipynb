# Task 9: Average Weekly Response Time for each agent

```python
def time_secs(time_str):
    hh, mm, ss =time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)

average_response_sec =[]
for t in df['Average Response Time']:
    average_response_sec.append(time_secs(t))
average_response_sec
df['average_response_sec']= average_response_sec
df
```

```python
df.groupby(['Agent Name', 'date_week'])['average_response_sec'].mean()
```

# Task 10: Average Weekly Resolution Time for each agent

```python
average_resolution_time=[]
for t in df['Average Resolution Time']:
    average_resolution_time.append(time_secs(t))
average_resolution_time

df['average_resolution_time']= average_resolution_time
df
```

```python
df.groupby(['Agent Name','date_week'])['average_resolution_time'].mean()
```

# Task 13 . Total contributation hour for each and every agents weekly basis

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1['week'] = df1['Date'].dt.week
```

```python
Total_duration_secs=[]
for t in df1['Duration']:
    Total_duration_secs.append(time_secs(t))
Total_duration_secs
df1['Total_duration_secs'] = Total_duration_secs
```
